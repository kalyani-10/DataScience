Stemming

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
```

```python
#create an object of class PorterStemmer
porter = PorterStemmer()
lancaster=LancasterStemmer()
#proide a word to be stemmed
print("Porter Stemmer")
print(porter.stem("friendships"))
print(porter.stem("friend"))
print(porter.stem("friendship"))
print(porter.stem("friends"))


print("Lancaster Stemmer")
print(lancaster.stem("friendships"))
print(lancaster.stem("friend"))
print(lancaster.stem("friendship"))
print(lancaster.stem("friends"))


```

```python

from nltk.tokenize import word_tokenize
ps = PorterStemmer()

sentence = "Pythoners are very intelligent and work very pythonly and now they are pythoning their way to success"
words = word_tokenize(sentence)

for w in words:
    print(w, " : ", ps.stem(w))
```

```python
#Lemmatizing sentence
from nltk.stem import WordNetLemmatizer
sentence = "the boy's cars are different colors, best go went gone going run runs ran and running"
words = nltk.word_tokenize(sentence)
for w in words:
    print(w, " : ", WordNetLemmatizer().lemmatize(w))

# pural to singular conversion
```

```python
#stop words removal
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

sentence ="Like cloud on cloud.No stir of air was there"
stop_words = set(stopwords.words('english'))

word_tokens = word_tokenize(sentence)

filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

filtered_sentence = []

for w in word_tokens:
	if w not in stop_words:
		filtered_sentence.append(w)

print(word_tokens)
print(filtered_sentence)

```
